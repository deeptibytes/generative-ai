#### Prompt Chain With Langchain And Open AI

In will demo how we will:

- Get setup with LangChain, LangSmith and LangServe
- Create a prompt chain and invoke llm
- Parse llm output with parser
- LangChain Expression Language (LCEL)

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

```python
## Create llm Langchain OpenAI Chat model
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model="gpt-4o")
print(llm)
```

```python
### Chatprompt Template
from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are an expert Java Engineer. Provide me answers based on the questions"),
        ("user","{input}")
    ]

)
prompt
```

```python
## chain 
chain=prompt|llm

response=chain.invoke({"input":"Can you tell me about lambdas in Java?"})
type(response)
print(response)
```

```python
## stroutput Parser

from langchain_core.output_parsers import StrOutputParser
output_parser=StrOutputParser()
chain=prompt|llm|output_parser

response=chain.invoke({"input":"Can you tell me about Stream API in Java?"})
print(response) 
```
